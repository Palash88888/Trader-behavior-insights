```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
pd.read_csv("historical_data.csv")
```

```python
pd.read_csv("fear_greed_index.csv")
```

```python
trader_df = pd.read_csv("historical_data.csv")
sentiment_df = pd.read_csv("fear_greed_index.csv")
```

```python
print(trader_df.columns.tolist())
```

```python
print(sentiment_df.columns.tolist())
```

```python
trader_df.columns = trader_df.columns.str.strip()
sentiment_df.columns = sentiment_df.columns.str.strip()
```

```python
trader_df['Timestamp IST'] = pd.to_datetime(trader_df['Timestamp IST'], errors='coerce')
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'], errors='coerce')
```

```python
trader_df['Date'] = trader_df['Timestamp IST'].dt.date
sentiment_df['Date'] = sentiment_df['date'].dt.date
```

```python
merged_df = pd.merge(trader_df, sentiment_df, on='Date', how='left')
```

```python
print(merged_df.head())
```

```python
merged_df = merged_df.dropna(subset=['classification'])
```

```python
sentiment_summary = merged_df.groupby('classification')['Closed PnL'].agg(['count', 'mean', 'sum']).reset_index()
```

```python
sentiment_summary.columns = ['Sentiment', 'Trade Count', 'Average PnL', 'Total PnL']
```

```python
print(sentiment_summary)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=sentiment_summary, x='Sentiment', y='Average PnL', palette='coolwarm')
plt.title('Average Closed PnL by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Closed PnL')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=sentiment_summary, x='Sentiment', y='Total PnL', palette='viridis')
plt.title('Total Closed PnL by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Total Closed PnL')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=sentiment_summary, x='Sentiment', y='Trade Count', palette='pastel')
plt.title('Number of Trades by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Number of Trades')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(data=sentiment_summary, x='Sentiment', y='Average PnL', marker='o', color='orange')
plt.title('Trend of Average PnL by Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average PnL')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(sentiment_summary['Trade Count'], labels=sentiment_summary['Sentiment'],
        autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Trade Distribution by Sentiment')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df, x='classification', y='Closed PnL', palette='coolwarm')
plt.title('Closed PnL Distribution by Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Closed PnL')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
corr = merged_df[['sentiment_score', 'Closed PnL']].corr()
print(corr)
```
